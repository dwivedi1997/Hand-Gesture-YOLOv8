```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from google.colab.patches import cv2_imshow
```

```python
img_name='/content/drive/MyDrive/Deep Learning/Object Detection/Hand-Gesture-Recognition-6/train/images/IMG-20230312-WA0065_jpg.rf.465771d5e2525d1245add62b20dac28e.jpg'
img_labels="/content/drive/MyDrive/Deep Learning/Object Detection/Hand-Gesture-Recognition-6/train/labels/IMG-20230312-WA0065_jpg.rf.465771d5e2525d1245add62b20dac28e.txt"
img=cv2.imread(img_name)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
label=pd.read_csv(img_labels,
                sep=' ',
                header=None,
                names=['label','x','y','w','h']
                  )
label
```

```python
img.shape
```

```python
color = (0, 255, 0)  # BGR color (green in this case)
thickness = 2
x=label.x
y=label.y
box_w=label.w
box_h=label.h
img_w=img.shape[1]
img_h=img.shape[0]
x1 = int((x - box_w / 2) * img_w)
y1 = int((y - box_h / 2) * img_h)
x2 = int((x + box_w / 2) * img_w)
y2 = int((y + box_h / 2) * img_h)
cv2.rectangle(img, (x1, y1), (x2,y2), color, thickness)
cv2_imshow(img)
```
